```python
#importing important libraries
# Data manipulation
import pandas as pd
import numpy as np

# Train-test split
from sklearn.model_selection import train_test_split

# Preprocessing 
from sklearn.preprocessing import StandardScaler

# Base models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    GradientBoostingClassifier,
    AdaBoostClassifier,
    ExtraTreesClassifier
)
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Evaluation metrics
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
df = pd.read_excel("OneDrive/Desktop/4th sem/DVZ/Book1.xlsx")

#Exploratory Data Analysis
print(df.shape)
df.head()
```

```python
#Identifying empty values
print(df.info())
print("\n")
print(df.isnull().sum())
```

```python
#Filling empty values
cols_to_fill = [
    'Value_Deal',
    'Multiple_Lines',
    'Online_Security',
    'Online_Backup',
    'Device_Protection_Plan',
    'Premium_Support',
    'Streaming_TV',
    'Streaming_Movies',
    'Streaming_Music',
    'Unlimited_Data',
    'Churn_Category',
    'Churn_Reason'
]

for col in cols_to_fill:
    df[col] = df[col].fillna('NA')
```

```python
#verification
print(df.isnull().sum())
```

```python
#label encoding of Customer Status
print(df['Customer_Status'].value_counts())
```

```python
print(df['Customer_Status'].unique())

df['Customer_Status'] = df['Customer_Status'].map({
    'Stayed': 0,
    'Joined': 0,
    'Churned': 1
})

print(df['Customer_Status'].value_counts())
```

```python
#exclude textual data which does not need one hot encoding
df = df.drop(
    columns=[
        'Churn_Category',
        'Churn_Reason'
    ]
)
```

```python
#set X and y such that in X, exclude Customer_ID(does not need one hot encoding) and Customer Status (target, assigned to y)
X = df.drop(
    columns=[
        'Customer_Status',
        'Customer_ID'
    ]
)

y = df['Customer_Status']
```

```python
#one hot encoding
X = pd.get_dummies(
    X,
    drop_first=True #reduce redundancy
)
```

```python
#divide the dataset into training data and testing data
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y #both sets preserve original distribution
)
```

```python
#grouping numerical columns
numeric_cols = [
    'Age',
    'Number_of_Referrals',
    'Tenure_in_Months',
    'Monthly_Charge',
    'Total_Charges',
    'Total_Refunds',
    'Total_Extra_Data_Charges',
    'Total_Long_Distance_Charges',
    'Total_Revenue'
]
```

```python
#standardize the dataset
scaler = StandardScaler()

X_train = X_train.copy()
X_test = X_test.copy()

X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=5000),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Extra Trees": ExtraTreesClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB()
}

results = []

for name, model in models.items():

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    y_prob = model.predict_proba(X_test)[:, 1]

    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),      # Correct predictions out of total predictions
        "Precision": precision_score(y_test, y_pred),    # Predicted positives that are actually positive
        "Recall": recall_score(y_test, y_pred),          # Actual positives correctly identified
        "F1 Score": f1_score(y_test, y_pred),            # Balance between precision and recall
        "ROC-AUC": roc_auc_score(y_test, y_prob)         # Ability to distinguish between the two classes
    })

results_df = pd.DataFrame(results)

results_df["Overall Score"] = (
    results_df["Accuracy"] +
    results_df["Precision"] +
    results_df["Recall"] +
    results_df["F1 Score"] +
    results_df["ROC-AUC"]
) / 5

best_model = results_df.sort_values(
    by="Overall Score",
    ascending=False
).iloc[0]

print(results_df.sort_values(by="Overall Score", ascending=False))

print("Chosen Model:", best_model["Model"])
print("Overall Score:", best_model["Overall Score"])
```

```python
# Train the chosen Logistic Regression model
model = LogisticRegression(max_iter=5000, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]
```

```python
#Class labels: 0 - Not Churned, 1 - Churned
print("Classes:", model.classes_)
```

```python
# Keep only active customers
active_df = df[df["Customer_Status"] == 0].copy()
```

```python
# Save Customer IDs for the final output
customer_ids = active_df["Customer_ID"]
```

```python
# Create prediction dataset
predict_df = active_df.drop(
    columns=[
        "Customer_ID",
        "Customer_Status"
    ]
)
```

```python
#ensuring that the prediction dataset has the same columns and order as the original dataset used to train the model
predict_df = predict_df.reindex(
    columns=X.columns,
    fill_value=0
)
```

```python
predict_df[numeric_cols] = scaler.transform(
    predict_df[numeric_cols]
)
```

```python
#Churn Probability prediction
churn_prob = model.predict_proba(predict_df)[:, 1]
```

```python
#Assigning risk
def risk(prob):

    if prob >= 0.60:
        return "High"

    elif prob >= 0.30:
        return "Medium"

    else:
        return "Low"
```

```python
risk_labels = [
    risk(x)
    for x in churn_prob
]
```

```python
final_df = pd.DataFrame({
    "Customer_ID": customer_ids.values
})
```

```python
final_df['Churn_Probability'] = (
    churn_prob * 100
).round(2)

final_df['Risk_Category'] = risk_labels
```

```python
high_risk = (
    final_df['Risk_Category']
    == 'High'
).sum()

print(
    "High Risk Customers:",
    high_risk
)
```
