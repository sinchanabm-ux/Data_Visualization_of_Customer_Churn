```python
#importing important libraries
# Data manipulation
import pandas as pd
import numpy as np

# Train-test split
from sklearn.model_selection import train_test_split

# Preprocessing 
from sklearn.preprocessing import StandardScaler

# Base models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    GradientBoostingClassifier,
    AdaBoostClassifier,
    ExtraTreesClassifier
)
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Evaluation metrics
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
df = pd.read_csv("Book1.csv")

print(df.shape)
df.head()
```

```python
#Identifying empty values
print(df.info())
print("\n")
print(df.isnull().sum())
```

```python
#Filling empty values
cols_to_fill = [
    'Value_Deal',
    'Multiple_Lines',
    'Online_Security',
    'Online_Backup',
    'Premium_Support',
    'Streaming_TV',
    'Streaming_Movies',
    'Streaming_Music',
    'Unlimited_Data',
    'Churn_Category',
    'Churn_Reason'
]

for col in cols_to_fill:
    df[col] = df[col].fillna('NA')
```

```python
#verification
print(df.isnull().sum())
```

```python
print(df['Customer_Status'].value_counts())
```

```python
#label encoding of Customer Status
print(df['Customer_Status'].unique())

df['Customer_Status'] = df['Customer_Status'].map({
    'Stayed': 0,
    'Joined': 0,
    'Churned': 1
})

print(df['Customer_Status'].value_counts())
```

```python
#exclude textual data which does not need one hot encoding
df = df.drop(
    columns=[
        'Churn_Category',
        'Churn_Reason'
    ]
)
```

```python
#set X and y such that in X, exclude Customer_ID(does not need one hot encoding) and Customer Status (target, assigned to y)
X = df.drop(
    columns=[
        'Customer_Status',
        'Customer_ID'
    ]
)

y = df['Customer_Status']
```

```python
#one hot encoding
X = pd.get_dummies(
    X,
    drop_first=True
)
```

```python
#divide the dataset into training data and testing data
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
#grouping numerical columns & standardize the dataset
numeric_cols = [
    'Age',
    'Number_of_Referrals',
    'Tenure_in_Months',
    'Monthly_Charge',
    'Total_Charges',
    'Total_Refunds',
    'Total_Long_Distance_Charges',
    'Total_Revenue'
]

scaler = StandardScaler()

X_train[numeric_cols] = scaler.fit_transform(
    X_train[numeric_cols]
)

X_test[numeric_cols] = scaler.transform(
    X_test[numeric_cols]
)
```

```python
#selecting the best model
models = {
    "Logistic Regression": LogisticRegression(max_iter=5000),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Extra Trees": ExtraTreesClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB()
}

results = []

for name, model in models.items():

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    y_prob = model.predict_proba(X_test)[:, 1]

    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),      # Correct predictions out of total predictions
        "Precision": precision_score(y_test, y_pred),    # Predicted positives that are actually positive
        "Recall": recall_score(y_test, y_pred),          # Actual positives correctly identified
        "F1 Score": f1_score(y_test, y_pred),            # Balance between precision and recall
        "ROC-AUC": roc_auc_score(y_test, y_prob)         # Ability to distinguish between the two classes
    })

results_df = pd.DataFrame(results)

results_df["Overall Score"] = (
    results_df["Accuracy"] +
    results_df["Precision"] +
    results_df["Recall"] +
    results_df["F1 Score"] +
    results_df["ROC-AUC"]
) / 5

best_model = results_df.sort_values(
    by="Overall Score",
    ascending=False
).iloc[0]

print(results_df.sort_values(by="Overall Score", ascending=False))

print("Chosen Model:", best_model["Model"])
print("Overall Score:", best_model["Overall Score"])
```

## Churn Risk Prediction

Score **only active customers** (Stayed + Joined).  
Churned customers are excluded — their churn has already occurred.

Risk thresholds (raw probability):
- **High Risk**: Churn Probability ≥ 0.50
- **Medium Risk**: 0.25 ≤ Churn Probability < 0.50
- **Low Risk**: Churn Probability < 0.25

```python
# Reload original data to get Customer_Status labels before encoding
raw_df = pd.read_csv("Book1.csv")

# Fill the same NaN columns
for col in cols_to_fill:
    raw_df[col] = raw_df[col].fillna('No')

# Keep only active customers (Stayed + Joined)
active_df = raw_df[raw_df['Customer_Status'].isin(['Stayed', 'Joined'])].copy()

print("Active customers to score:", len(active_df))
print(active_df['Customer_Status'].value_counts())
```

```python
# Drop columns not used as features
predict_df = active_df.drop(
    columns=[
        'Customer_ID',
        'Customer_Status',
        'Churn_Category',
        'Churn_Reason'
    ]
)

# One-hot encode to match training features
predict_df = pd.get_dummies(predict_df, drop_first=True)

# Align columns exactly with training set (handles unseen categories)
predict_df = predict_df.reindex(columns=X.columns, fill_value=0)

# Scale numeric columns using the SAME scaler fitted on training data
predict_df[numeric_cols] = scaler.transform(predict_df[numeric_cols])
```

```python
# Get churn probability (raw decimal, column index for 'Churned' class)
churn_idx = list(model.classes_).index(1)
churn_prob = model.predict_proba(predict_df)[:, churn_idx]
```

```python
def risk(prob):
    """
    Classify churn risk using raw probability thresholds.
    High   : prob >= 0.50
    Medium : 0.25 <= prob < 0.50
    Low    : prob < 0.25
    """
    if prob >= 0.50:
        return "High Risk"

    elif prob >= 0.25:
        return "Medium Risk"

    else:
        return "Low Risk"

risk_labels = [risk(p) for p in churn_prob]
```

```python
# Build the output dataframe
final_df = pd.DataFrame({
    'Customer_ID':        active_df['Customer_ID'].values,
    'Churn_Probability':  (churn_prob * 100).round(2),   # stored as %
    'Risk_Category':      risk_labels
})

# Summary
print("Risk Category Distribution:")
print(final_df['Risk_Category'].value_counts())
print()
print("Sample — top 20 highest risk:")
final_df.sort_values('Churn_Probability', ascending=False).head(20)
```
